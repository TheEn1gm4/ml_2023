```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
xi=np.array([1, 1.6, 3.4, 4.0, 5.1])
yi=np.array([1.2, 2.0, 2.4, 3.5, 3.5])
#PROCEDIMIENTO
```

```python
a, b= np.polyfit(xi, yi, 1)
r= np.corrcoef(xi, yi) 
```

```python
plt.plot(xi, yi, 'o')
plt.xlim(np.min(xi)-1, np.max(xi)+1)
plt.ylim(np.min(yi)-1, np.max(yi)+1)
plt.plot(xi, a * xi + b)
```
